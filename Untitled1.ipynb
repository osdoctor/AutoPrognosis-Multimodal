```python
import numpy as np, pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python

# -----------------------------
# (A) Synthetic data generation (n=200)
# -----------------------------
rng = np.random.default_rng(42)
n = 200
true_cluster = rng.choice(4, size=n, p=[0.28, 0.26, 0.24, 0.22])

def clip_int(x, lo, hi):
    return np.clip(np.rint(x).astype(int), lo, hi)

def gen_range_codes(c):
    # cranial/caudal-most level codes (1~15 used)
    if c == 0:
        cran = rng.integers(2, 5); caud = cran + rng.integers(6, 10)   # longer
    elif c == 1:
        cran = rng.integers(4, 7); caud = cran + rng.integers(3, 6)    # mid
    elif c == 2:
        cran = rng.integers(6, 9); caud = cran + rng.integers(5, 8)    # lower extension
    else:
        cran = rng.integers(3, 8); caud = cran + rng.integers(6, 10)   # wide/complex
    caud = int(np.clip(caud, 1, 15))
    cran = int(np.clip(cran, 1, caud))
    return cran, caud

rows = []
for i in range(n):
    c = int(true_cluster[i])
    cran, caud = gen_range_codes(c)
    span = caud - cran + 1

    # phenotype-driven morphology
    if c == 0:
        fragments = clip_int(rng.normal(1.3, 0.6), 1, 6)
        att_bone  = clip_int(rng.normal(min(7, span*0.7), 1.0), 0, 7)
        att_disc  = clip_int(rng.normal(min(6, span*0.25), 0.8), 0, 6)
        flo_bone  = clip_int(rng.normal(0.4, 0.6), 0, 7)
        flo_disc  = clip_int(rng.normal(0.2, 0.5), 0, 6)
        cor       = int(rng.choice([2,3,4], p=[0.25,0.45,0.30]))
    elif c == 1:
        fragments = clip_int(rng.normal(2.6, 0.9), 1, 8)
        att_bone  = clip_int(rng.normal(min(7, span*0.35), 1.2), 0, 7)
        att_disc  = clip_int(rng.normal(min(6, span*0.55), 1.1), 0, 6)
        flo_bone  = clip_int(rng.normal(0.7, 0.8), 0, 7)
        flo_disc  = clip_int(rng.normal(0.4, 0.7), 0, 6)
        cor       = int(rng.choice([1,2,3], p=[0.15,0.50,0.35]))
    elif c == 2:
        fragments = clip_int(rng.normal(2.2, 0.9), 1, 8)
        att_bone  = clip_int(rng.normal(min(7, span*0.25), 1.2), 0, 7)
        att_disc  = clip_int(rng.normal(min(6, span*0.25), 1.0), 0, 6)
        flo_bone  = clip_int(rng.normal(min(7, span*0.55), 1.2), 0, 7)
        flo_disc  = clip_int(rng.normal(min(6, span*0.45), 1.1), 0, 6)
        cor       = int(rng.choice([1,2,3], p=[0.20,0.55,0.25]))
    else:
        fragments = clip_int(rng.normal(3.0, 1.2), 1, 10)
        att_bone  = clip_int(rng.normal(min(7, span*0.45), 1.3), 0, 7)
        att_disc  = clip_int(rng.normal(min(6, span*0.35), 1.2), 0, 6)
        flo_bone  = clip_int(rng.normal(min(7, span*0.35), 1.2), 0, 7)
        flo_disc  = clip_int(rng.normal(min(6, span*0.30), 1.0), 0, 6)
        cor       = int(rng.choice([2,3,4], p=[0.35,0.40,0.25]))

    # clinical (for later interpretation; not used for clustering below)
    age = float(np.round(np.clip(rng.normal(62, 10), 30, 90), 1))
    sex_male = int(rng.choice([0,1], p=[0.35,0.65]))
    preop_joa = float(np.round(np.clip(14 - 1.2*(cor-1) - 0.15*span + rng.normal(0, 1.5), 0, 17), 0))

    rows.append({
        "patient_id": f"P{(i+1):03d}",
        "age": age,
        "sex_male": sex_male,
        "preop_joa": preop_joa,
        "attached_bone_levels": int(att_bone),
        "attached_disc_levels": int(att_disc),
        "floating_bone_levels": int(flo_bone),
        "floating_disc_levels": int(flo_disc),
        "cranial_most_level_code": int(cran),
        "caudal_most_level_code": int(caud),
        "total_opll_fragments": int(fragments),
        "max_cor_grade": int(cor),
    })

df = pd.DataFrame(rows)

# -----------------------------
# (B) Unsupervised clustering (feature-table 기반)
# -----------------------------
morph_cols = [
    "attached_bone_levels","attached_disc_levels",
    "floating_bone_levels","floating_disc_levels",
    "cranial_most_level_code","caudal_most_level_code",
    "total_opll_fragments","max_cor_grade"
]

X = StandardScaler().fit_transform(df[morph_cols])

# 데모에서는 k=4로 고정 (실제 연구에서는 silhouette/Gap/statistical stability로 결정 권장)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=20).fit(X)
df["cluster"] = kmeans.labels_

# cluster 요약
summary = df.groupby("cluster")[morph_cols + ["age","sex_male","preop_joa"]].mean().round(2)
print(df["cluster"].value_counts().sort_index())
print(summary)

# 저장
df.to_csv("synthetic_opll_200.csv", index=False)
df.to_excel("synthetic_opll_200.xlsx", index=False)
```

### Explanation of Synthetic Data Generation

The synthetic dataset was generated to simulate patient data related to Ossification of the Posterior Longitudinal Ligament (OPLL) for clustering analysis. Here's a breakdown of its generation process:

1.  **Sample Size:** The dataset comprises `n = 200` synthetic patient samples.

2.  **Morphological Features:** These features describe the characteristics of the OPLL lesion and were used for clustering. They include:
    *   `attached_bone_levels`: Number of bone levels to which the OPLL is attached.
    *   `attached_disc_levels`: Number of disc levels to which the OPLL is attached.
    *   `floating_bone_levels`: Number of floating bone levels of OPLL (not attached to adjacent structures).
    *   `floating_disc_levels`: Number of floating disc levels of OPLL.
    *   `cranial_most_level_code`: The code for the cranial-most spinal level affected by OPLL.
    *   `caudal_most_level_code`: The code for the caudal-most spinal level affected by OPLL.

### Data Preprocessing and K-Means Clustering for OPLL Subtyping

To prepare the morphological features for clustering, a crucial data preprocessing step was performed:

1.  **Feature Selection**: Eight morphological features, identified as key descriptors of OPLL lesions, were selected for clustering. These features include `attached_bone_levels`, `attached_disc_levels`, `floating_bone_levels`, `floating_disc_levels`, `cranial_most_level_code`, `caudal_most_level_code`, `total_opll_fragments`, and `max_cor_grade`.

2.  **Standardization**: Before applying the clustering algorithm, these eight morphological features were standardized. Standardization is essential because clustering algorithms, especially K-Means, are sensitive to the scale of the features. Features with larger ranges could disproportionately influence the distance calculations, leading to biased clustering results. The `StandardScaler` from `sklearn.preprocessing` was used for this purpose. This transformation scales each feature such that it has a mean of 0 and a standard deviation of 1. The standardized data was stored in a matrix named `X`.

Following preprocessing, the K-Means clustering algorithm was applied:

1.  **Algorithm**: K-Means clustering, an unsupervised learning algorithm, was used to group patients into distinct clusters based on their morphological similarities.

2.  **Parameters**: The following parameters were specified for the K-Means algorithm:
    *   `n_clusters=4`: For this demonstration, the number of clusters was fixed at 4. In a real-world research scenario, the optimal number of clusters would typically be determined using methods like the silhouette score, Gap statistic, or statistical stability analysis.
    *   `random_state=42`: This parameter ensures reproducibility. By setting a fixed `random_state`, the initial placement of centroids is consistent across multiple runs, yielding the same clustering results each time.
    *   `n_init=20`: The algorithm was run 20 times with different centroid seeds. K-Means clustering is sensitive to the initial placement of centroids, and running it multiple times with different initializations helps in finding a more robust and globally optimal clustering solution. The best result out of these 20 runs (in terms of inertia, i.e., sum of squared distances of samples to their closest cluster center) was chosen.

3.  **Cluster Assignment**: The resulting cluster labels, obtained from the K-Means model, were then assigned back to the original `df` DataFrame as a new column named `'cluster'`. This allows for easy integration and analysis of cluster assignments alongside the original patient and morphological data.

#### Instructions
### Dimensionality Reduction using PCA and Cluster Visualization

To visually inspect the separation of the identified clusters, Principal Component Analysis (PCA) was employed for dimensionality reduction, followed by a scatter plot visualization.

1.  **Principal Component Analysis (PCA)**:
    *   The standardized morphological feature matrix `X` was subjected to PCA. PCA is a linear dimensionality reduction technique that transforms correlated variables into a smaller number of uncorrelated variables called principal components, while retaining as much of the original variance as possible.
    *   The `PCA` class from `sklearn.decomposition` was used.
    *   The data was reduced to two principal components (`n_components=2`) to allow for 2D visualization.
    *   A `random_state=42` was set to ensure reproducibility of the component extraction.
    *   The `fit_transform` method was applied to `X` to compute the principal components and project the data onto this new 2-dimensional space, resulting in `X_pca`.

2.  **Cluster Visualization via Scatter Plot**:
    *   The two principal components (`X_pca[:, 0]` and `X_pca[:, 1]`) were used as the x and y axes for a scatter plot.
    *   Each data point in the scatter plot was colored according to its assigned cluster label, which was stored in the `df['cluster']` column.
    *   The `seaborn.scatterplot` function was utilized for its enhanced aesthetic capabilities.
    *   A `viridis` color palette was applied to distinguish the different clusters.
    *   The plot was titled 'PCA of Morphological Features by Cluster' and appropriate labels were added for 'Principal Component 1' and 'Principal Component 2'.
    *   A legend was included to map colors to cluster identities, and a grid was added for better readability. This visualization allows for qualitative assessment of cluster separation in a lower-dimensional space.
